**Calculating distribution (<font color="red">Methods-Q4</font> - 3 pts)**: Describe how the percentage of mitochondrial genes and number of counts are calculated from the matrix to add to the observations for the AnnData object. What were the top 3 barcodes of the cells that had the highest percentage of mitochondrial genes? What were the top 3 barcodes of the cells that had the highest number of counts?

The code below works through how the percentage of mitochondrial genes and number of counts are calculate from the matrix in orderd to add to the AnnData object as observations. First the genes that start with 'mt-' are identified. Then for each cell the sum of all counts of mitochondrial genes are summed together and then divided by the total counts per gene. This gives what percent of counts of genes for a single cell are mitochondrial. This is then added to the obs dataframe (adata.obs['percent_mito']). Lastly all of the gene counts per cell are summed (axis = 1 is across the rows) to give the total number of gene counts per cell. This is added to the observation dataframe as well (adata.obs['n_counts']).

````python
import numpy as np
mito_genes = adata.var_names.str.startswith('mt-')
# for each cell compute fraction of counts in mito genes vs. all genes
# the `.A1` is only necessary as X is sparse (to transform to a dense array after summing)
adata.obs['percent_mito'] = np.sum(adata[:, mito_genes].X, axis=1).A1 / np.sum(adata.X, axis=1).A1
## add the total counts per cell as observations-annotation to adata
sc.pp.filter_cells(adata, min_genes=0)
adata.obs['n_counts'] = adata.X.sum(axis=1).A1

# You can sort by a column using e.g. adata.obs.sort_values(colname, ascending=False) to get highest values at the top
````

## Discussion

**<font color="red">Discussion-Q1</font> - 3 pts**: Describe the differences between 3'-end based scRNA-seq technologies like 10X Genomics vs. full transcript methods like SMART-seq. What are advantages/disadvantages of each, and for which applications would each one be preferred?

The 3'-end based scRNA-seq does not give full insight into the different regions of the genes. For example we loose isophorm data because we do not know where the introns/exons are when using the 3 primed ends. Also we do not get information from the beginning of transcripts such as mutations. Furthermore when using full transcript methods like SMART-seq we get more full transcripts from the individual cells, but we get a lot fewer cells (lower throughput) in general since they are done in individual wells. A lot more effort is needed in generating that data. It becomes a tradeoff of quality vs quantity when comparing the two methods. The 10X Genomics method may be preferred when the samples that we are analyzing are heterogeneous and you want to see how the transcript is distributed among different cells in a sample. SMART-seq may be preferred in the situation that you already have purified cells all of a single type. Since there is only a single type of cell we want to get the most information of those cell's expression rather than comparing to other cells.
